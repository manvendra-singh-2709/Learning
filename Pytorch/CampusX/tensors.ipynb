# Creating Tensor

```python
et = torch.empty(2, 3)
print(et)
```

```python
z = torch.zeros(2, 3)
print(z)
```

```python
o = torch.ones(2, 3)
print(o)
```

```python
r = torch.rand(2, 3)
print(r)
```

```python
torch.manual_seed(42)
rs = torch.rand(2, 3)
print(rs)
```

```python
t = torch.tensor([[1 , 2, 3],[5, 6, 7]])
print(t)
```

```python
print("using  arange-> ", torch.arange(0, 10, 2))

print("using linspace -> ", torch.linspace(1, 10 ,10))

print("using eye -> ", torch.eye(5))

print("using full -> ", torch.full((3, 3), 5))
```
